```python
def ob_cols(df):
    cols = df.columns
    ob_cols = []
    for i in range(len(cols)):
        if df[cols[i]].dtypes == 'object':
            ob_cols.append(cols[i])
    print(ob_cols)
```

```python
df1['feature_9'] = df1['feature_9'].astype('float')
```

```python
df1['engine_has_gas'] = df1['engine_has_gas'].astype('float')
df1['has_warranty'] = df1['has_warranty'].astype('float')
```

```python
var_dums2 = ['manufacturer_name','color','body_type']
dums2 = pd.get_dummies(df1[var_dums2],drop_first=True)
df1.drop(var_dums2,axis=1,inplace=True)
```

```python
df1 = pd.concat([df1,dums2],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X = df1.drop('price_usd',axis=1).values
y = df1['price_usd'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
y_train = y_train.reshape(-1,1)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(79, activation="relu"))
model.add(Dense(40, activation="relu"))
model.add(Dense(20, activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
#model.fit(x=X_train, y=y_train, epochs=30, verbose=1)
```

```python
history = model.history.history
plt.plot(history['loss'])
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,preds)
print(mse**0.5)
```
